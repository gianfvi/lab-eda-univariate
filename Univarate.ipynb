```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
print(df.head())
```

```python
category_counts = df['category'].value_counts()

top_5_categories = category_counts.head(5)
print("Top 5 Most Listed Product Categories:\n", top_5_categories)

```

```python

# Visualization: Distribution of products across different categories

plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Listings')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization: Proportions of top categories using a pie chart
top_categories = category_counts.head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.title('Proportion of Top 10 Product Categories')
plt.show()
```

```python
# Part 2: Delving into Product Pricing
# Calculate measures of centrality for product price
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
```

```python
print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

```python
price_variance = df['price'].var()
price_std_dev = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)
```

```python
print(f"Variance of Price: {price_variance}")
print(f"Standard Deviation of Price: {price_std_dev}")
print(f"Range of Price: {price_range}")
print(f"Interquartile Range of Price: {price_iqr}")
```

```python
# Visualization: Histogram of product prices
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()
```

```python
# Visualization: Box plot of product prices
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['price'])
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Part 3: Unpacking Product Ratings
# Calculate measures of centrality for product ratings
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]
```

```python

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")
```

```python
rating_variance = df['stars'].var()
rating_std_dev = df['stars'].std()
rating_iqr = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)
```

```python
print(f"Variance of Rating: {rating_variance}")
print(f"Standard Deviation of Rating: {rating_std_dev}")
print(f"Interquartile Range of Rating: {rating_iqr}")
```

```python
rating_skewness = df['stars'].skew()
rating_kurtosis = df['stars'].kurt()

print(f"Skewness of Rating: {rating_skewness}")
print(f"Kurtosis of Rating: {rating_kurtosis}")
```

```python
# Visualization: Histogram of product ratings
plt.figure(figsize=(12, 6))
sns.histplot(df['stars'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```
