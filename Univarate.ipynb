```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
file_path = 'amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(file_path)

# Part 1: Understanding Product Categories
# Business Question: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

# Frequency Tables
category_freq = df['category'].value_counts()
top_5_categories = category_freq.head(5)

print("Frequency Table for Product Categories:")
print(category_freq)
print("\nTop 5 Most Listed Product Categories:")
print(top_5_categories)

# Visualizations
# Bar chart for product categories
plt.figure(figsize=(12, 6))
category_freq.plot(kind='bar', color='skyblue')
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Listings')
plt.show()

# Pie chart for top 5 categories
plt.figure(figsize=(8, 8))
top_5_categories.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Proportion of Top 5 Product Categories')
plt.ylabel('')
plt.show()



```

```python
# Part 2: Delving into Product Pricing
# Business Question: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

# Measures of Centrality
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print("\nMeasures of Centrality for Product Prices:")
print(f"Mean: {mean_price}")
print(f"Median: {median_price}")
print(f"Mode: {mode_price}")

# Measures of Dispersion
price_variance = df['price'].var()
price_std_dev = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print("\nMeasures of Dispersion for Product Prices:")
print(f"Variance: {price_variance}")
print(f"Standard Deviation: {price_std_dev}")
print(f"Range: {price_range}")
print(f"Interquartile Range: {price_iqr}")

# Visualizations
# Histogram for product prices
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], kde=True, bins=30, color='purple')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box plot for product prices
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], color='orange')
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Part 3: Unpacking Product Reviews
# Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

# Measures of Centrality
mean_review = df['reviews'].mean()
median_review = df['reviews'].median()
mode_review = df['reviews'].mode()[0]

print("\nMeasures of Centrality for Product Reviews:")
print(f"Mean: {mean_review}")
print(f"Median: {median_review}")
print(f"Mode: {mode_review}")

# Measures of Dispersion
review_variance = df['reviews'].var()
review_std_dev = df['reviews'].std()
review_iqr = df['reviews'].quantile(0.75) - df['reviews'].quantile(0.25)

print("\nMeasures of Dispersion for Product Reviews:")
print(f"Variance: {review_variance}")
print(f"Standard Deviation: {review_std_dev}")
print(f"Interquartile Range: {review_iqr}")

# Shape of the Distribution
review_skewness = skew(df['reviews'].dropna())
review_kurtosis = kurtosis(df['reviews'].dropna())

print("\nShape of the Distribution for Product Reviews:")
print(f"Skewness: {review_skewness}")
print(f"Kurtosis: {review_kurtosis}")

# Visualizations
# Histogram for product reviews
plt.figure(figsize=(12, 6))
sns.histplot(df['reviews'], kde=True, bins=10, color='green')
plt.title('Distribution of Product Reviews')
plt.xlabel('Reviews')
plt.ylabel('Frequency')
plt.show()
```
